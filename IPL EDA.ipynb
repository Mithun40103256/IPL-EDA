```python
# Importing necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing data
df = pd.read_csv('matches.csv')
df.head(3)
```

```python
del(df['id'])
```

```python
df.head(3)
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(15,8))
df['winner'].value_counts().plot(kind='bar',colormap='summer',linewidth=3.5,edgecolor='k')
plt.xlabel('Teams',fontweight='bold',fontsize=15)
plt.ylabel('Number of matches won',fontweight='bold',fontsize=15)
plt.title('NUMBER OF MATCHES WON BY A TEAM',fontsize=20,fontweight='bold',style = 'italic')
plt.xticks(rotation=80);
```

```python
plt.figure(figsize=(15,10))
df['player_of_match'].value_counts()[:20].plot(kind='bar',colormap='Paired',linewidth=3.5,edgecolor='k')
plt.xlabel('Number of man of the match',fontweight='bold',fontsize=15)
plt.ylabel('Name of the players',fontweight='bold',fontsize=15)
plt.title('MAN OF THE MATCH (TOP 20)',fontsize=20,fontweight='bold',style = 'italic')
plt.xticks(rotation=80);
```

```python
df
```

```python
plt.figure(figsize=(15,10))
df['city'].value_counts().plot(kind='bar',colormap='Accent',linewidth=3.5,edgecolor='k')
plt.xlabel('City name',fontweight='bold',fontsize=15)
plt.ylabel('Number of matches played',fontweight='bold',fontsize=15)
plt.title('VENUE',fontsize=20,fontweight='bold',style = 'italic')
plt.xticks(rotation=80);
```

```python
toss_win_match_win = df[df['toss_winner']==df['winner']]
```

```python
print(f"{len(toss_win_match_win)} times a team won the toss won the match and {len(df)-len(toss_win_match_win)} times a team won the toss did not win the match")
```

```python
len(toss_win_match_win)/len(df)
```

## 51.1% of the time a team won the toss managed to win the match

```python
toss_win_match_win['toss_decision'].value_counts()
```

```python
201/325,124/325
```

## 61.84% of the time a team won the match by winning the toss decided to field first and 38.16% of the time they decided to bat first

```python
plt.figure(figsize=(15,8))
toss_win_match_win['win_by_runs'].plot(kind='hist',colormap='Spectral',linewidth=3.5,edgecolor='k')
plt.title('WIN MARGIN BY RUNS OF THE TEAM WHO DECIDED TO BAT FIRST',fontsize=22,fontweight='bold',style='italic')
plt.xlabel('Winning margin by runs',fontsize=16)
plt.ylabel('Frequency',fontsize=16)
```

```python
plt.figure(figsize=(15,8))
toss_win_match_win['win_by_wickets'].plot(kind='hist',colormap='PuOr',linewidth=3.5,edgecolor='k')
plt.title('WIN MARGIN BY WICKETS OF THE TEAM WHO DECIDED TO BOWL FIRST',fontsize=22,fontweight='bold',style='italic')
plt.xlabel('Winning margin by wickets',fontsize=16)
plt.ylabel('Frequency',fontsize=16)
```

## From the above 2 figures we can say that all the matches were very competitive.Most of the times teams managed to win by 1 wicket or by around 10 runs
